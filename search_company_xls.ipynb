```python
# !/usr/bin/python
# -*- coding: utf-8 -*-
# Author: Selvaria
# 在序时账内查找指定岛外企业

import os

doclist = {}
com_name = ''

def gci(filepath):
    global com_name
    # 遍历filepath下所有文件，包括子目录
    files = os.listdir(filepath)
    for fi in files:
#         if fi == '.DS_Store':
#             continue
        fi_d = os.path.join(filepath, fi)
        if os.path.isdir(fi_d):
            if '公司' in fi:
                com_name = fi
                doclist[com_name] = []
            gci(fi_d) #如果有子文件的递归
        else:
            doclist[com_name].append(fi_d)
    return doclist

path_list = gci('D:\\Selvaria\\scripts\\company\\changxing_invoice/XSZ')
xml = []
excel = []
excel_other = []
for k, v in path_list.items():
    # print(k, v)
    if not v:
        continue
    for path in v:
#         if '序时' in path or '余额' in path:
#             continue
        if 'xml' in path:
            xml.append([k, path])
        else:
            if 'xls' in path or 'xlsx' in path:
                excel.append([k, path])
            else:
                excel_other.append([k, path])
```

```python
import pandas
import xlrd

com_in_list = [] #包含待查询的企业列表
com_error = [] #文件结构错误
path_error_list = [] #文件格式错误
# name = '山东海科石化销售有限公司'
name = '粤海石化（深圳）有限公司|深圳市申钢贸易有限公司'

def get_df(path, name):
    data = xlrd.open_workbook(path, encoding_override="gb18030")  # 打开xls文件
    table = data.sheets()[0]
    record = []
    nrows = table.nrows  # 获取表的行数
    for i in range(nrows):
        if i < 1:  # 跳过第一行
            continue
        record.append(table.row_values(i))
    length = len(record[0])
    
    df2 = pandas.DataFrame(record,columns=list(range(length)))
#     print(length)
#     各类文件结构
#     if length == 12:
#         df = df2[df2[2].str.contains(name)] 
#     elif length == 10:
#         df = df2[df2[2].str.contains(name)]
#     elif length == 11:
#         df = df2[df2[5].str.contains(name)]
#     elif length == 27:
#         df = df2[df2[8].str.contains(name)]
#     elif length == 28:
#         df = df2[df2[8].str.contains(name)]

    if length == 28:
        df = df2[df2[8].str.contains(name)] 
    elif length == 27:
        df = df2[df2[8].str.contains(name)]
    else:
        return -2
    result = len(df)
#     result = df
    return result

for l in excel:
    try:
        res = get_df(l[1], name)
    except:
        print('企业格式出现问题：', l)
        path_list.append(l)
        continue
    if res>0:
        com_in_list.append(l[0])
    elif res == -2:
        #print(l[0])
        com_error.append(l)
    else:        
        continue
```
