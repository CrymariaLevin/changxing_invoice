```python
# 开始插入新的无id产品
import pandas

df_insert = pandas.read_excel('d:/无分类id产品20200526.xls')
df_insert.head(30)
```

```python
df_str = df_insert.copy()
df_str['main_id'] = df_str['main_id'].apply(lambda x: '0'+ str(int(x)) if len(str(int(x)))<6 else str(int(x)))
df_str['spec_id'] = df_str['spec_id'].apply(lambda x: '0'+ str(int(x)) if len(str(int(x)))<6 else str(int(x)))
df_str['good_id'].fillna('220100', inplace=True)
df_str['good_id'] = df_str['good_id'].apply(lambda x: '0'+ str(int(x)) if len(str(int(x)))<6 else str(int(x)))
# df_str['good_id'] = df_str['good_id'].apply(lambda x: '' if x=='00' else x)
df_str['full_name'].fillna('', inplace=True)
df_str['good_name'].fillna('', inplace=True)
df_str.head(30)
```

```python
data = df_str[['exchange_good', 'main_id', 'spec_id', 'good_id', 'full_name']].values
data = [list(data_info) for data_info in data]
data
```
